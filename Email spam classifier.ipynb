```python
import os
from collections import Counter
```

```python
folder='emails/'
```

```python
files = os.listdir(folder)
```

```python
len(files)
```

```python
emails = [folder + file for file in files]
```

```python
words = []

for email in emails:
    f = open(email, encoding='latin-1')
    blob = f.read()
    words+=blob.split(" ")
```

```python
len(words)
```

```python
for i in range(len(words)):
    if not words[i].isalpha():
        words[i] =""
```

```python
word_dict = Counter(words)
```

```python
word_dict
```

```python
len(word_dict)
```

```python
del word_dict[""]
```

```python
word_dict
```

```python
word_dict = word_dict.most_common(3000)
```

```python
word_dict
```

```python
features = []
labels = []
for email in emails:
    f = open(email,encoding='latin-1')
    blob=f.read().split(" ")
    data = []
    for i in word_dict:
        data.append(blob.count(i[0]))
    features.append(data)
    
    if 'spam' in email:
        labels.append(1)
    if 'ham' in email:
        labels.append(0)
```

```python
len(features)
```

```python
len(labels)
```

```python
import numpy as np
```

```python
features = np.array(features)
features.shape
```

```python
labels = np.array(labels)
labels.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(features, labels, test_size=0.2, random_state=9)
```

```python
x_train.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
```

```python
classifier.fit(x_train,y_train)
```

```python
new_email = """you have unlocked a new offer. 
deposit money in order  to claim  the prize. hurry up before
the stock is empty"""
```

```python
sample = []
for i in word_dict:
    sample.append(new_email.split(" ").count(i[0]))
```

```python
sample = np.array(sample)
```

```python
len(sample)
```

```python
classifier.predict(sample.reshape(1,3000))
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)
```
